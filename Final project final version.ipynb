```python
#Numerical packages
import numpy as np
import pandas as pd
#--------------------------------------------
#Statistic package
from sklearn.linear_model import LinearRegression
from scipy.stats import shapiro
import statsmodels.api as sm
#--------------------------------------------
#Graphical packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pylab
```

```python
df_data = pd.read_csv('~/Desktop/AirQuality7.csv')
df_data.head()
```

```python
model = LinearRegression()
y = np.array(df_data['CO'].values)
alpha = 0.05
```

# Benceno

```python
comp = 'Benceno'

x = np.array(df_data[comp].values).reshape(-1,1)

values_model_benceno = model.fit(x,y)

print('Intercept:', values_model_benceno.intercept_)
print('Slope:', values_model_benceno.coef_)
print('Determination coefficient:', values_model_benceno.score(x, y))

determination_coeff_Benceno = values_model_benceno.score(x, y)

CO_prediccion_Benceno = model.predict(x)

df_data.plot(x=comp, y='CO', style='o')
plt.plot(x, CO_prediccion_Benceno)
plt.title('Benceno vs CO')
plt.xlabel('Benceno')
plt.ylabel('CO')
plt.show()

Benceno_Residuos_prediccion = CO_prediccion_Benceno-y
Benceno_Residuos_estandarizado = (Benceno_Residuos_prediccion-Benceno_Residuos_prediccion.mean())/Benceno_Residuos_prediccion.std()

plt.hist(Benceno_Residuos_estandarizado)
plt.title('Distribucion de residuos de la prediccion de CO usando '+comp)
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.show()
#Shapiro-Wilk Test
W_benceno, pvalue_benceno = shapiro(Benceno_Residuos_estandarizado)
print('Estadistico W=%.4f, valor-p=%.4f' % (W_benceno, pvalue_benceno))
if pvalue_benceno > alpha:
    print('No rechazar H0, existe evidencia estadistica para aceptar que los datos siguen una distribucion normal')
else:
    print('Rechazar H0, no existe evidencia estadistica para aceptar que los datos siguen una distribucion normal')
#q-q plot    
sm.qqplot(Benceno_Residuos_estandarizado, line='45')
pylab.show()
```
